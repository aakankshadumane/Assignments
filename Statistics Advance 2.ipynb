```python
import numpy as np
from scipy.stats import norm,poisson
import matplotlib.pyplot as plt
```

```python
# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? 
# Explain with an example.
1. Probability Mass Function (PMF) : Function that describes probability of different discrete
    values of a variable.
    ex. Rolling a dice has outcomes 1,2,3,4,5,6
    Each value has equal probability hence pmf is (1/6)= 0.167
    
2. Probability Density Function (PDF) : Function that describes probability of different continuos
    values of a variable.
    ex . height of people
    It can have values like 165,165.3,165.5,166 cms
    When plotted, it gives normally distributed curve,
```

```python
# PDF
data = np.random.normal(100,5,500) #mean,standard_deviation and size given as arguments
print("Probablity at value 100 is :",norm.pdf(100,100,5))
#Plot probability density histogram
plt.hist(data,density=True)  
# Provided X and Y labels
plt.xlabel("data values")
plt.ylabel("Probabilities")
```

```python
# PMF
data = np.arange(0, 100, 1) #Array of discrete values
y = poisson.pmf(data, mu=10, loc=40)  
plt.plot(data, y)
plt.xlabel("Data values")
plt.ylabel("Probablities")
```

```python
# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
It is a function that describes the distribution of random variables.It can be defined for 
any kind of random variable (discrete, continuous, and mixed).
CDF is used to determine the likelihood that a random observation taken from the population will 
be less than or equal to a particular value.
```

```python
Example: 
data = np.random.normal(100,5,500) #mean,standard_deviation and size given as arguments
print("Probablity at value 100 is :",norm.cdf(100,100,5))

# Calculate the cumulative proportion of the data that falls below each value
cumulative = np.linspace(0, 1, len(data))

# Sort the data in ascending order
sorted_data = np.sort(data)

# Calculate the cumulative proportion of the sorted data
cumulative_data = np.cumsum(sorted_data) / np.sum(sorted_data)

#Plot cdf
plt.plot(sorted_data, cumulative_data)  
# Provided X and Y labels
plt.xlabel("data values")
plt.ylabel("Probabilities")
```

```python
# Q3: What are some examples of situations where the normal distribution might be used as a model?
# Explain how the parameters of the normal distribution relate to the shape of the distribution.
Examples:
    1. Height of a Population: most people have the average height and fewer people ar at the 
        extremes (very tall or very short).
    2. IQ Scores: IQ scores in the general population do follow a normal distribution, with an 
        average IQ of 100 and a standard deviation of 15. 
    3. Test Scores: In educational testing, the scores on standardized tests like the SAT or GRE 
        often approximate a normal distribution. 
        
Two parameters relate to shape of distribution:
1. Mean : It is at the center of the normal distribution.When the mean is increased, the 
    distribution shifts to the right, and when it is decreased, the distribution shifts to the left.
2. Standard Deviation : A larger standard deviation (σ) results in a wider distribution with more variability.
A smaller standard deviation (σ) results in a narrower distribution with less variability.
```

```python
# Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal
# Distribution.
Importance of Normal/Guassian Distribution:
1. Common Data Approximation: The normal distribution has a reasonable approximation for the distribution 
    of many real-world phenomena.
2. Statistical Inference: Many statistical methods and hypothesis tests assume that the data follows a normal 
    distribution. These methods include t-tests, ANOVA, and regression analysis.
3. Comparative Analysis: When comparing different groups or populations, the normal distribution provides 
    a standard reference point.

Real-life Examples:
1. Birth Weights: The birth weights of newborn babies are often normally distributed, with most infants 
    having weights close to the average, and fewer babies having extremely high or low birth weights.
2. Blood Pressure: Most individuals have blood pressure values close to the mean, while fewer have 
    extremely high or low blood pressure readings.
3. Quality Control: In manufacturing, the distribution of defects in a product or the number of 
    defects per unit can be modeled using a normal distribution. 
```

```python
# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
# Distribution and Binomial Distribution?
The Bernoulli distribution is a discrete probability distribution that models a random experiment 
with only two possible outcomes: success and failure.It is often used to represent simple events 
like flipping a coin, rolling a die to check for a specific outcome, or success/failure in a 
binary experiment.

Example:Coin Flip
Consider a fair coin as our random experiment. We can use the Bernoulli distribution to model 
the outcome of a single coin flip.

In this case:
A "success" (X = 1) represent getting heads (H).
A "failure" (X = 0) represent getting tails (T).
In this Bernoulli distribution:

The probability of success (getting heads) is p = 0.5 because it's a fair coin.
The probability of failure (getting tails) is 1 - p = 0.5.

Difference :
1. Number of Trials:
Bernoulli: The Bernoulli distribution models a single trial or experiment with only two 
    possible outcomes: success or failure.
Binomial: The binomial distribution models the number of successes in a fixed number of 
    independent and identical Bernoulli trials.It deals with the number of successes in a sequence of trials.
    
2.Parameters:
Bernoulli : The Bernoulli distribution has a single parameter, denoted as "p," which represents 
    the probability of success in a single trial. The probability of failure is 1 - p.
Binomial : The binomial distribution has two parameters: "n" (the number of trials) and "p" 
        (the probability of success in each trial).
```

```python
# Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that 
# the dataset is normally distributed, what is the probability that a randomly selected 
# observation will be greater than 60? Use the appropriate formula and show your calculations.
Given : mean = 50, standard deviation = 10 , Normally Distributed
1. Calculate the z-score for the value 60
z = (X - mean) / SD
z = (60 - 50) / 10 = 1

2. observation is greater than 60, which corresponds to the area to the right of the z-score 
of 1 on the standard normal distribution curve.
P(Z < 1) is  0.8413.
P(X > 60) = 1 - P(Z < 1)

3.Calculate the final probability:
P(X > 60) = 1 - 0.8413 ≈ 0.1587
Hence, probablity is 0.1587 or 15.87%.
```

```python
# Q7: Explain uniform Distribution with an example.
The uniform distribution is a continuous probability distribution in which all outcomes within a 
given range are equally likely. It is characterized by a constant probability density function (PDF) 
across this range. Every value within the interval has the same probability of occurring.
The uniform distribution is not limited to discrete values; it can also be continuous, where the 
values within a continuous range have a constant probability density.

Example: Rolling a Fair Die
Consider the experiment of rolling a fair die. In this case, each face of the die has an equal probability 
of 1/6 of landing face up. The outcomes are the numbers 1, 2, 3, 4, 5, and 6.
P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = 1/6
```

```python
# Q8: What is the z score? State the importance of the z score.
Z score is a statistical measure that quantifies how far a given data point or observation is 
from the mean of a dataset in terms of standard deviations.
z = (X - mean) / SD

Importance:
1. Standardization: It transforms data into a common scale where the mean is 0 and the standard 
    deviation is 1. 
2. Identification of Outliers: A z-score significantly greater than or less than zero indicates 
    that a data point is far from the mean and may be considered an outlier.
3. Hypothesis Testing:  In hypothesis tests like the z-test, z-scores are used to assess whether 
    a sample mean is significantly different from a population mean.
```

```python
# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
Central Limit Theorem states that Sampling distribution of means will always be normally 
distributed as long as sample size is large enough, regardless of the shape of the original 
population distribution.


Significance:
1. Normal Distribution Approximation: The CLT allows us to approximate the sampling distribution
    of the sample mean as a normal distribution, even when the population distribution itself is not normal.
2. Wider Applicability:  It enables statisticians and data scientists to apply these methods even when the 
    population data doesn't follow a normal distribution. 
3. Inference About Population Parameters : It allows us to make inferences about population parameters, 
    such as the population mean, based on sample data.
```

```python
# Q10: State the assumptions of the Central Limit Theorem.
Key Assumptions: 
The samples are randomly and independently drawn from the population.
The population has a finite mean (μ) and a finite standard deviation (σ).
The samples are large enough (n > 30)
The CLT does not require the population to be normally distributed. However, it is 
important to note that the larger the sample size, the less important this assumption becomes.
```
